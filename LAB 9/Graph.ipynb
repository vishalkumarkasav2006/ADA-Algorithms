```python
# All_pairs_shortest_path

import matplotlib.pyplot as plt
import numpy as np

# Replace with your actual output
n_values = [20, 40, 60, 80, 100]
time_values = [1, 1, 4, 8, 16]  # example

n_arr = np.array(n_values)

# Theoretical curves
y_n = n_arr
y_n2 = n_arr ** 2
y_n3 = n_arr ** 3
y_logn = np.log2(n_arr)

# Scaling
scale_n = time_values[-1] / y_n[-1]
scale_n2 = time_values[-1] / y_n2[-1]
scale_n3 = time_values[-1] / y_n3[-1]
scale_logn = time_values[-1] / y_logn[-1]

y_n_sc = y_n * scale_n
y_n2_sc = y_n2 * scale_n2
y_n3_sc = y_n3 * scale_n3
y_logn_sc = y_logn * scale_logn

# Plot
plt.figure(figsize=(12, 7))

plt.plot(n_values, time_values, 'ro-', linewidth=3, label='Floyd-Warshall')
plt.plot(n_values, y_n3_sc, 'g--', label='O(V^3)')
plt.plot(n_values, y_n2_sc, 'b--', label='O(V^2)')
plt.plot(n_values, y_n_sc, 'orange', linestyle='dotted', label='O(V)')
plt.plot(n_values, y_logn_sc, 'purple', linestyle='dotted', label='O(log V)')

plt.title('Floyd-Warshall vs Theoretical Complexities')
plt.xlabel('Number of Vertices (V)')
plt.ylabel('Time (ms)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
