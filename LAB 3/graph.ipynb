```python
### for the selection sort 

import matplotlib.pyplot as plt
import numpy as np
n_values =[10000,20000,25000,50000,75000,100000,150000,200000]
time_values =[9,38,37,75,113,149,216,295] 
n_arr = np.array(n_values)
y_n = n_arr                 
y_nsquare= n_arr ** 2            
y_logn = np.log2(n_arr)        
y_nlogn = n_arr * np.log2(n_arr)
scale_n = time_values[-1] / y_n[-1]
scale_nsquare= time_values[-1] / y_nsquare[-1]
scale_logn = time_values[-1] / y_logn[-1]
scale_nlogn = time_values[-1] / y_nlogn[-1]
y_n_sc = y_n * scale_n
y_nsquare_sc = y_nsquare * scale_nsquare
y_logn_sc = y_logn * scale_logn
y_nlogn_sc = y_nlogn * scale_nlogn
plt.figure(figsize=(12, 7))
# Actual Data (Thick Red Line)
plt.plot(n_values, time_values,'ro-', linewidth=3, label='Actual MERGE Sort')
plt.plot(n_values, y_nsquare_sc,'b--', label='O(N^2) Quadratic')
plt.plot(n_values, y_nlogn_sc, 'g--', alpha=0.5, label='O(N log N)')
plt.plot(n_values, y_n_sc,   'orange', linestyle='dotted', alpha=0.5, label='O(N) Linear')
plt.plot(n_values, y_logn_sc, 'purple', linestyle='dotted', alpha=0.5, label='O(log N)')
plt.title(f'MERGE SORT vs Theoretical Complexities')
plt.xlabel('Input Size (N)')
plt.ylabel('Time (ms)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
