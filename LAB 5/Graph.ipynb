```python
# convex_hull

import matplotlib.pyplot as plt
import numpy as np

# Replace with your actual benchmark values
n_values = [1000, 5000, 10000, 20000, 50000 ]
time_values = [0, 2, 4, 10, 25]   # example (ms)

n_arr = np.array(n_values)
# Theoretical curves
y_n = n_arr
y_nsquare = n_arr ** 2
y_logn = np.log2(n_arr)
y_nlogn = n_arr * np.log2(n_arr)

# Scaling (to match graph visually)
scale_n = time_values[-1] / y_n[-1]
scale_nsquare = time_values[-1] / y_nsquare[-1]
scale_logn = time_values[-1] / y_logn[-1]
scale_nlogn = time_values[-1] / y_nlogn[-1]
y_n_sc = y_n * scale_n
y_nsquare_sc = y_nsquare * scale_nsquare
y_logn_sc = y_logn * scale_logn
y_nlogn_sc = y_nlogn * scale_nlogn

# Plot
plt.figure(figsize=(12, 7))
plt.plot(n_values, time_values, 'ro-', linewidth=3, label='Convex Hull (Actual)')
plt.plot(n_values, y_nlogn_sc, 'g--', label='O(N log N)')
plt.plot(n_values, y_nsquare_sc, 'b--', label='O(N^2)')
plt.plot(n_values, y_n_sc, 'orange', linestyle='dotted', label='O(N)')
plt.plot(n_values, y_logn_sc, 'purple', linestyle='dotted', label='O(log N)')
plt.title('Convex Hull vs Theoretical Complexities')
plt.xlabel('Input Size (N)')
plt.ylabel('Time (ms)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
