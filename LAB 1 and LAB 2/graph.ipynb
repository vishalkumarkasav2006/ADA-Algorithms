```python
## best case insertion sort, 
import matplotlib.pyplot as plt
import numpy as np
n_values = [10, 100, 1000, 10000,100000, 1000000,1000000, 10000000]
time_values = [0, 0, 0,  0, 0, 2 , 16 , 254] 
n_arr = np.array(n_values)
y_n      = n_arr                
y_nlogn  = n_arr * np.log2(n_arr) 
y_n2     = n_arr ** 2   
scale_n     = time_values[-1] / y_n[-1]
scale_nlogn = time_values[-1] / y_nlogn[-1]
scale_n2    = time_values[-1] / y_n2[-1]
y_n_scaled      = y_n * scale_n
y_nlogn_scaled  = y_nlogn * scale_nlogn
y_n2_scaled     = y_n2 * scale_n2
plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, 'ro-', linewidth=3, label='Actual C++ Time')
plt.plot(n_values, y_n2_scaled, 'b--', alpha=0.5, label='O(N^2) - Quadratic')
plt.plot(n_values, y_nlogn_scaled, 'g--', alpha=0.5, label='O(N log N)')
plt.plot(n_values, y_n_scaled, 'orange', linestyle='--', alpha=0.5, label='O(N) - Linear')
plt.title(f'Actual Time vs Theoretical Growth (N up to {n_values[-1]})')
plt.xlabel('Input Size (N)')
plt.ylabel('Time (ms)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
### for the bubble sort 
import matplotlib.pyplot as plt
import numpy as np
n_values =[1000, 5000,10000,25000,50000,75000,100000]
time_values =[0, 20, 89,577,2278, 5139,9408] 
n_arr = np.array(n_values)
y_n = n_arr                 
y_nsquare= n_arr ** 2            
y_logn = np.log2(n_arr)        
y_nlogn = n_arr * np.log2(n_arr)

scale_n = time_values[-1] / y_n[-1]
scale_nsquare= time_values[-1] / y_nsquare[-1]
scale_logn = time_values[-1] / y_logn[-1]
scale_nlogn = time_values[-1] / y_nlogn[-1]
y_n_sc = y_n * scale_n
y_nsquare_sc = y_nsquare * scale_nsquare
y_logn_sc = y_logn * scale_logn
y_nlogn_sc = y_nlogn * scale_nlogn

plt.figure(figsize=(12, 7))
# Actual Data (Thick Red Line)
plt.plot(n_values, time_values, 'ro-', linewidth=3, label='Actual Bubble Sort')
# Theoretical Curves (Dashed Lines)
plt.plot(n_values, y_nsquare_sc, 'b--', label='O(N^2) Quadratic')
plt.plot(n_values, y_nlogn_sc,   'g--', alpha=0.5, label='O(N log N)')
plt.plot(n_values, y_n_sc,       'orange', linestyle='dotted', alpha=0.5, label='O(N) Linear')
plt.plot(n_values, y_logn_sc,    'purple', linestyle='dotted', alpha=0.5, label='O(log N)')
plt.title(f'Bubble Sort vs Theoretical Complexities')
plt.xlabel('Input Size (N)')
plt.ylabel('Time (ms)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
## permute code hmm, 
import matplotlib.pyplot as plt
import numpy as np
import math
n_values=[8,  9,   10,    11 , 12 , 13]
time_values=[4,28,230, 2085,21556, 287087] 
n_arr = np.array(n_values)
y_fact = np.array([math.factorial(n) for n in n_values]) 
y_exp  = 2.0 ** n_arr                                    
y_n2   = n_arr ** 2                                      
scale_fact = time_values[-1] / y_fact[-1]
scale_exp  = time_values[-1] / y_exp[-1]
scale_n2   = time_values[-1] / y_n2[-1]
y_fact_sc = y_fact * scale_fact
y_exp_sc  = y_exp * scale_exp
y_n2_sc   = y_n2 * scale_n2
plt.figure(figsize=(12, 7))
plt.plot(n_values, time_values, 'ro-', linewidth=4, label='Actual Permutation Time')
plt.plot(n_values, y_fact_sc, 'purple', linestyle='--', linewidth=2, label='O(N!) - Factorial (The Match)')
plt.plot(n_values, y_exp_sc,  'b--', alpha=0.5, label='O(2^N) - Exponential')
plt.plot(n_values, y_n2_sc,   'g--', alpha=0.3, label='O(N^2) - Quadratic')
plt.title(f'Permutation Generation vs Factorial Growth')
plt.xlabel('String Length (N)')
plt.ylabel('Time (ms)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
