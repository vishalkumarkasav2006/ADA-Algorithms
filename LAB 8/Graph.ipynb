```python
# Matrix chain multiplication

import matplotlib.pyplot as plt
import numpy as np

# Replace with your actual output
n_values = [100, 200, 300, 400, 500]
time_values = [2, 23, 80, 189, 371]  # example
n_arr = np.array(n_values)

# Theoretical curves
y_n = n_arr
y_n2 = n_arr ** 2
y_n3 = n_arr ** 3
y_logn = np.log2(n_arr)

# Scaling
scale_n = time_values[-1] / y_n[-1]
scale_n2 = time_values[-1] / y_n2[-1]
scale_n3 = time_values[-1] / y_n3[-1]
scale_logn = time_values[-1] / y_logn[-1]
y_n_sc = y_n * scale_n
y_n2_sc = y_n2 * scale_n2
y_n3_sc = y_n3 * scale_n3
y_logn_sc = y_logn * scale_logn

# Plot
plt.figure(figsize=(12, 7))
plt.plot(n_values, time_values, 'ro-', linewidth=3, label='MCM (Memo DP)')
plt.plot(n_values, y_n3_sc, 'g--', label='O(N^3)')
plt.plot(n_values, y_n2_sc, 'b--', label='O(N^2)')
plt.plot(n_values, y_n_sc, 'orange', linestyle='dotted', label='O(N)')
plt.plot(n_values, y_logn_sc, 'purple', linestyle='dotted', label='O(log N)')
plt.title('Matrix Chain Multiplication vs Theoretical Complexities')
plt.xlabel('Number of Matrices (N)')
plt.ylabel('Time (ms)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
